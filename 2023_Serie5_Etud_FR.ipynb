## Exercice 1
Ecrivez un programme qui interprête une liste d'instructions en RPN (Reverse Polish Notation).
* Si l'instruction est un entier, ajoutez-la au sommet de la pile.
* Les opérations + - * / déclenchent un `pop()` des deux derniers nombres de la pile et un `push()` du résultat de l'opération entre ces deux nombres.

Exemples:

- ```1 4 - 3 *``` donne comme résultat ```-9```

- ```3 4 * 6 / 1 2 +``` donne comme résultat ```2 3```

Votre solution doit utiliser une pile que vous avez implémentée. Elle doit utiliser une liste (tableau) de taille fixe sans pour autant hériter de `List`. Il faut utiliser les fonctions (que vous devez implémenter) suivantes:

````python
class MyStack:
    def __init__(self, size: int):
    # initialize the stack with a fixed size and set top to -1 (empty stack)
        # A COMPLETER
        # pass
        self.stack = [0] * size
        self.top = -1

    def print(self):
    # print the elements in the stack
        # A COMPLETER
        # pass
        for i in range(self.top + 1):
            print(self.stack[i], end=" ")
        print()

    def size(self):
    # return the current size of the stack
        # A COMPLETER
        # return None # retourne un int
        return self.top + 1

    def is_empty(self):
    # check if the stack is empty (top is -1)
        # A COMPLETER
        # return None # retourne un boolean
        return self.top == -1

    def push(self, o: int): # Throws FullStackException
    # add an element to the stack (if not full)
        # A COMPLETER
        # pass
        if self.top < len(self.stack) - 1:
          self.top += 1
          self.stack[self.top] = o
        else:
          raise FullStackException("Error: Stack is full")

    def pop(self): # Throws VoidStackException
    # remove and return the top element from the stack (if not empty)
        # A COMPLETER
        # return None # retourne un int
        if not self.is_empty():
            value = self.stack[self.top]
            self.top -= 1
            return value
        else:
            raise VoidStackException("Error: Stack is empty")

class Error(Exception):
    pass

class FullStackException(Error):
    pass

class VoidStackException(Error):
    pass
````

````python
s = MyStack(3)
assert s.size() == 0
assert s.is_empty() == True
s.push(1)
s.push(4)
s.print()
assert s.size() == 2
assert s.is_empty() == False
assert s.pop() == 4
s.print()
assert s.size() == 1
s.push(3)
s.push(10)
assert s.size() == 3
s.print()
try:
    s.push(12)
    print("Erreur: FullStackException doit être levée durant cette opération")
except FullStackException:
    pass
except:
    print("Erreur: FullStackException doit être levée durant cette opération")
assert s.pop() == 10
assert s.pop() == 3
assert s.pop() == 1
assert s.size() == 0
assert s.is_empty() == True

s = MyStack(3)
s.push(5)
assert s.size() == 1
assert s.is_empty() == False
s.print()
assert s.pop() == 5
assert s.size() == 0
assert s.is_empty() == True
s.print()
try:
    s.pop()
    print("Erreur: VoidStackException doit être levée durant cette opération")
except VoidStackException:
    pass
except:
    print("Erreur: VoidStackException doit être levée durant cette opération")
````

````python
# Param:
# entry: String[]
def rpn(entry):
# create an instance of the MyStack class to manage the stack
    # A COMPLETER
    stack = MyStack(len(entry))                     # create a stack with a fixed size
    operators = set(["+", "-", "*", "/"])           # define valid operators

    for token in entry:
        if token.isdigit() or (token[0] == '-' and token[1:].isdigit()):
            # if the token is a number, push it onto the stack
            stack.push(int(token))
        elif token in operators:
            # if the token is an operator, perform the operation on the top two values from the stack
            if stack.size() < 2:
                print("Error: Not enough operands for the operator", token)
                return None
            operand2 = stack.pop()
            operand1 = stack.pop()
            if token == "+":
                result = operand1 + operand2
            elif token == "-":
                result = operand1 - operand2
            elif token == "*":
                result = operand1 * operand2
            elif token == "/":
                if operand2 == 0:
                    print("Error: Division by zero")
                    return None
                result = operand1 / operand2
            stack.push(int(result))
        else:
            print("Error: Invalid instruction", token)
            return None

    # return the resulting stack containing the calculated values
    return stack
    # return None # retourne un objet MyStack
````

````python
s = rpn(["1","4","-","3","*"])
s.print() # doit être "-9"
s = rpn(["3","4","*","6","/","1","2","+"])
s.print() # doit être "2 3"
````

### Explications

The `rpn` function takes a list of instructions in Reverse Polish Notation (RPN) as its input. It creates an instance of the `MyStack` class, which is defined above, to manage the stack data structure. The size of the stack is set to be the same as the number of elements in the entry list. The function defines a set of valid operators which are (+, -, *, /) and will be used in the RPN calculations.

The function then iterates through each token in the entry list. If the token is a number, either positive or negative, it is converted to an integer and pushed onto the stack using the `push` method of the `MyStack` class. If the token is an operator, it checks if there are at least two operands on the stack. If not, it prints an error message and returns `None`. If there are enough operands, it pops the top two values from the stack and performs the corresponding operation based on the operator. The result is then pushed back onto the stack. If the token is neither a number nor a valid operator, it prints an error message and returns `None`.

After processing all the tokens in the entry list, the function returns the stack, which now contains the calculated result of the RPN expression. In summary, this function evaluates an RPN expression by iteratively processing each token in the input list, performing required calculations using a stack, and returning the final result as a stack. It handles errors such as insufficient operands and division by zero.

````python
class DoublyLinkedList:
    head = None  # Tête de la liste
    last = None  # Dernier élément de la liste

    def __init__ (self):
        self.head = Node()  # Crée un nœud sentinelle pour la tête de la liste
        self.last = self.head  # Initialise le dernier élément de la liste avec la tête

    def is_empty(self):
        return self.head == self.last  # Renvoie vrai si la liste est vide

    def begin(self):
        if self.is_empty():
            return DoublyLinkedIterator(self.head)  # Renvoie un itérateur pointant vers la tête si la liste est vide
        return DoublyLinkedIterator(self.head.get_next())  # Renvoie un itérateur pointant vers le premier élément de la liste

    def end(self):
        return DoublyLinkedIterator(self.last)  # Renvoie un itérateur pointant vers le dernier élément de la liste

    def add(self, e):
        new_last = Node(self.last, e, None)  # Crée un nouveau nœud avec la valeur 'e'
        self.last.set_next(new_last)  # Met à jour le lien 'next' du dernier nœud pour pointer vers le nouveau nœud
        self.last = new_last  # Met à jour le dernier élément de la liste pour pointer vers le nouveau nœud

    def remove(self):
        if self.is_empty():
            return None  # Renvoie None si la liste est vide
        first_node = self.head.get_next().get()  # Récupère la valeur du premier élément
        if self.head.get_next() == self.last:  # Si la liste contient un seul élément
            self.head.set_next(None)  # Met à jour les liens pour indiquer que la liste est vide
            self.last = self.head
        else:
            self.head.get_next().get_next().set_previous(self.head)  # Met à jour les liens pour contourner le premier nœud
            self.head.set_next(self.head.get_next().get_next())  # Met à jour le lien 'next' de la tête
        return first_node  # Renvoie la valeur du premier élément supprimé

class DoublyLinkedIterator:
    crt_node = None  # Noeud courant

    def __init__(self, node):
        self.crt_node = node  # Initialise l'itérateur avec un nœud initial

    def set(self, e):
        if self.crt_node:
            self.crt_node.set(e)  # Modifie la valeur de l'élément actuellement itéré

    def get(self):
        if self.crt_node:
            return self.crt_node.get()  # Renvoie la valeur stockée dans l'élément actuellement itéré

    def increment(self):
        if self.crt_node:
            return DoublyLinkedIterator(self.crt_node.get_next())  # Passe à l'élément suivant s'il existe

    def decrement(self):
        if self.crt_node:
            return DoublyLinkedIterator(self.crt_node.get_previous())  # Passe à l'élément précédent s'il existe

    def equals(self, other):
        if isinstance(other, DoublyLinkedIterator):
            if self.crt_node and other.crt_node:
                return self.crt_node == other.crt_node  # Renvoie vrai si les deux itérateurs sont sur le même élément
            return self.crt_node is None and other.crt_node is None
        return False  # Renvoie faux si les deux itérateurs ne parcourent pas la même liste

class Node:
    previous_node = None  # Noeud précédent
    current_node = None  # Noeud courant
    next_node = None  # Noeud suivant

    def __init__(self, previous_node=None, current_node=None, next_node=None):
        self.current_node = current_node  # Initialise le nœud avec une valeur courante
        self.previous_node = previous_node  # Initialise le lien vers le nœud précédent
        self.next_node = next_node  # Initialise le lien vers le nœud suivant

    def get(self):
        return self.current_node  # Renvoie la valeur stockée dans le nœud

    def set(self, node):
        self.current_node = node  # Modifie la valeur stockée dans le nœud

    def get_next(self):
        return self.next_node  # Renvoie le nœud suivant

    def set_next(self, node):
        self.next_node = node  # Met à jour le lien vers le nœud suivant

    def get_previous(self):
        return self.previous_node  # Renvoie le nœud précédent

    def set_previous(self, node):
        self.previous_node = node  # Met à jour le lien vers le nœud précédent
````
